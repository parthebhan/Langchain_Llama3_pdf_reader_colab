```python
from PyPDF2 import PdfReader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma

from langchain.schema import Document
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.llms import Ollama
```

```python
# Example PDF path in Google Drive (replace with your actual path)
pdf = '/content/drive/MyDrive/Colab Notebooks/llama_index_pdf_reader/ind_geo.pdf'
```

```python
def get_pdf_text(pdf_docs):
    text = ""
    for pdf in pdf_docs:
        pdf_reader = PdfReader(pdf)
        for page in pdf_reader.pages:
            text += page.extract_text()
    return text


def get_text_chunks(text):
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=10000, chunk_overlap=1000)
    chunks = text_splitter.split_text(text)
    return chunks
```

```python
text = get_pdf_text([pdf])
chunks = get_text_chunks(text)
```

# Vector Embeddings


```python
# Assuming 'chunks' is a list of strings that you want to convert into documents
documents = [Document(page_content=chunk) for chunk in chunks]
```

```python
embedding = OllamaEmbeddings(model="nomic-embed-text", show_progress=True)
embedding
```

```python
# Add to vector database
vector_db = Chroma.from_documents(
    documents=documents,
    embedding=embedding,
    collection_name="local-rag"
)
```

```python
vector_db
```

# Retrival


```python
from langchain.prompts import ChatPromptTemplate, PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_community.chat_models import ChatOllama
from langchain_core.runnables import RunnablePassthrough
from langchain.retrievers.multi_query import MultiQueryRetriever
```

```python
QUERY_PROMPT = PromptTemplate(
    input_variables=["question"],
    template="""You are an expert assistant with access to a vector database containing informations.
    Use the information from the database to answer the following query with 2000 words if you have.
    If you do not find sufficient information in the database to answer the query accurately, respond with "I don't have enough information to answer that."
    ### Query:{question}""",
)
```

```python
retriever = MultiQueryRetriever.from_llm(   vector_db.as_retriever(),
                                            llm,
                                            prompt=QUERY_PROMPT
                                        )

# RAG prompt
template = """Answer the question based ONLY on the following context:{context} Question: {question} """

prompt = ChatPromptTemplate.from_template(template)
```

```python
chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
chain.invoke(input(""))
```

```python
query = "national wildlife action plans in India"
chain.invoke(query)
```
